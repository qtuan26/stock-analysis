```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
import glob
import os

plt.style.use("seaborn-v0_8")
```

```python
PROC_DIR = "../data/processed"

files = glob.glob(os.path.join(PROC_DIR, "*.csv"))
tickers = [os.path.basename(f).replace(".csv", "") for f in files]

print("Tickers phát hiện:", tickers)
```

```python
def load_ticker(ticker):
    path = f"../data/processed/{ticker}.csv"
    df = pd.read_csv(path, parse_dates=["Date"])
    df = df.sort_values("Date").reset_index(drop=True)
    return df

# Test
df = load_ticker(tickers[0])
df.head()
```

```python
def plot_close(df, ticker):
    plt.figure(figsize=(12,4))
    plt.plot(df["Date"], df["Adj Close"], linewidth=1.5)
    plt.title(f"{ticker} – Close Price")
    plt.xlabel("Date")
    plt.ylabel("Close")
    plt.grid(True)
    plt.show()

plot_close(df, tickers[0])
```

```python
def plot_volume(df, ticker):
    plt.figure(figsize=(12,4))
    plt.bar(df["Date"], df["Volume"], width=1)
    plt.title(f"{ticker} – Volume")
    plt.show()

plot_volume(df, tickers[0])
```
